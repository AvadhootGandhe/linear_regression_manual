```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abhishek14398/salary-dataset-simple-linear-regression")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df=pd.read_csv(f"{path}/Salary_dataset.csv")
```

```python
df['Salary'].max()
```

```python
import matplotlib.pyplot as plt
```

```python
def gradient_decent(m_now,b_now,points,lr):
  m_gradient=0
  b_gradient=0
  n=len(points)
  for i in range(n):
    x=points.iloc[i].YearsExperience
    y=points.iloc[i].Salary
    m_gradient+=-(2/n)*(y-(m_now*x+b_now))*x
    b_gradient+=-(2/n)*(y-(m_now*x+b_now))
  m=m_now-m_gradient*lr
  b=b_now-b_gradient*lr
  return m,b
m=0
b=0
lr=0.0001
epochs=1000
for i in range(epochs):
  m,b=gradient_decent(m,b,df,lr)
```

```python
print(m,b)
```

```python
plt.scatter(df.YearsExperience,df.Salary)
```

```python
x_vals = list(range(0, 20))
y_vals = [m * x + b for x in x_vals]
plt.plot(x_vals, y_vals)
plt.scatter(df.YearsExperience,df.Salary)
plt.show()
```
